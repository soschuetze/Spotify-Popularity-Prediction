```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import json
```

```python
credentials = json.load(open('authorization.json'))
```

```python
client_id = credentials['client_id']
client_secret = credentials['client_secret']
```

```python
import numpy as np
import pandas as pd
```

```python
client_credentials_manager = SpotifyClientCredentials(client_id=client_id,client_secret=client_secret)
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)
```

```python
df = pd.DataFrame(columns=['name','album','artist','release_date','length','popularity','danceability','acousticness','energy','instrumentalness','liveness','loudness','speechiness','valence','tempo'])

for artist in ['drake','badbunny','theweeknd','edsheeran']:
    artist_name = artist
    with open(artist+'.json') as f:
        data = json.load(f)
        counter = 0
        for line in data['items']:
            album_uri = line['uri']
            album = sp.albums([album_uri], market='US')
            for a in album['albums']:
                counter+=1
                if counter%2==0:
                    pass
                else:
                    album = a['name']
                    for t in a['tracks']['items']:
                        release_date = line['release_date']
                        artist_name = artist
                        track_uri = t['uri']
                        track_info = sp.track(track_uri, market='US')
                        popularity = track_info['popularity']
                        name = t['name']
                        audio_features = sp.audio_features(track_uri)
                        for song in audio_features:
                            length = song['duration_ms']
                            energy = song['energy']
                            danceability = song['danceability']
                            acousticness = song['acousticness']
                            instrumentalness = song['instrumentalness']
                            liveness = song['liveness']
                            loudness = song['loudness']
                            speechiness = song['speechiness']
                            valence = song['valence']
                            tempo = song['tempo']
                            df.loc[len(df.index)] = [name, album, artist_name, release_date, length, popularity, danceability, acousticness, energy, instrumentalness, liveness, loudness, speechiness, valence, tempo] 
    f.close()
```
